# Gradient Descent on MSE
Simple Linear Equation: $$y = wx + b; w = weight, b = bias$$
Cost Function: $$MSE = \frac{1}{N} \sum({y-yprediction})^2$$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("student_scores.csv")
data.head()
```

```python
plt.scatter(data.Hours, data.Scores)
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.show()
```

```python
def gradient_descent(x, y, l_rate, epoch):
    w_curr = b_curr = 0
    n = len(x)
    
    for i in range(epoch):
        y_predicted = w_curr * x + b_curr
        
        #partial derivatives of MSE
        dw = -(2/n)*sum(x*(y-y_predicted))
        db = -(2/n)*sum(y-y_predicted)
        
        w_curr = w_curr - l_rate * dw
        b_curr = b_curr - l_rate * db
        
    return w_curr, b_curr
```

```python
x = data.Hours
y = data.Scores
```

```python
m, b = gradient_descent(x, y, 0.0001, 10000)
print(m, b)
```

```python
plt.scatter(x, y)
plt.plot(list(range(1, 10)), [m*x+b for x in range(1, 10)], color="red")
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.show()
```
