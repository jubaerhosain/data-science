```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
pd = df = pd.read_csv('../../../datasets/1_homeprices.csv')
df
```

Data Preprocessing: Fill NA values with median value of a column

```python
df.bedrooms.median()
```

```python
df.bedrooms = df.bedrooms.fillna(df.bedrooms.median())
df
```

```python
df.drop('price',axis='columns')
```

```python
lr_model = linear_model.LinearRegression()
lr_model.fit(df.drop('price',axis='columns'), df.price)
```

```python
lr_model.coef_
```

```python
lr_model.intercept_
```

```python
lr_model.predict([[3000, 3, 40]])
```
